```python
import yt_dlp
import http.cookiejar
import os


def download_youtube_video(youtube_url, save_path="video.mp4", cookies_path="cookies.txt"):

    cookiejar = http.cookiejar.MozillaCookieJar()
    if os.path.exists(cookies_path):
        cookiejar.load(cookies_path, ignore_discard=True, ignore_expires=True)
    else:

        with open(cookies_path, 'w') as f:
            pass

    ydl_opts = {
        'format': 'best',
        'outtmpl': save_path,
        'cookiefile': cookies_path
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

    return save_path
```

# **Downloading The YouTube Video**

```python
youtube_url = "https://www.youtube.com/watch?v=I-x0G5pAan0"
video_path = download_youtube_video(youtube_url)
print(f"Video downloaded: {video_path}")
```

# **Loading a YOLOv8 Model for Object Detection**

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

# **Real-Time Object Detection in the given Video**

```python
import cv2
from google.colab.patches import cv2_imshow

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break


    results = model(frame)

    for result in results:
        boxes = result.boxes.xyxy
        confidences = result.boxes.conf
        classes = result.boxes.cls

        for box, conf, cls in zip(boxes, confidences, classes):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[int(cls)]} {conf:.2f}"


            if model.names[int(cls)] in ["knife", "person"]:
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)


    cv2_imshow(frame)


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
