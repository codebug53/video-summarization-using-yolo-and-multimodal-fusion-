```python
import yt_dlp
import http.cookiejar
import os


def download_youtube_video(youtube_url, save_path="video.mp4", cookies_path="cookies.txt"):
    # Load cookies from file
    cookiejar = http.cookiejar.MozillaCookieJar()
    if os.path.exists(cookies_path):
        cookiejar.load(cookies_path, ignore_discard=True, ignore_expires=True)
    else:
        # If the cookies file doesn't exist, create an empty one.
        with open(cookies_path, 'w') as f:
            pass

    ydl_opts = {
        'format': 'best',
        'outtmpl': save_path,
        'cookiefile': cookies_path  # Pass the cookiefile path, not the cookiejar object
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])

    return save_path
```

```python
youtube_url = "https://www.youtube.com/watch?v=I-x0G5pAan0"
video_path = download_youtube_video(youtube_url)
print(f"Video downloaded: {video_path}")
```

```python
from ultralytics import YOLO

# Load the pre-trained YOLOv8 model
model = YOLO("yolov8n.pt")  # Change to 'yolov8s.pt' or a custom model for better accuracy
```

```python
# Open video file
import cv2 # Import the cv2 module (OpenCV)
from google.colab.patches import cv2_imshow # Import cv2_imshow for Colab

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # End of video

    # Run YOLOv8 object detection
    results = model(frame)

    # Draw detections on the frame
    for result in results:
        boxes = result.boxes.xyxy  # Bounding box coordinates
        confidences = result.boxes.conf  # Confidence scores
        classes = result.boxes.cls  # Detected class labels

        for box, conf, cls in zip(boxes, confidences, classes):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[int(cls)]} {conf:.2f}"

            # Detect people and knives (potential stabbing indicators)
            if model.names[int(cls)] in ["knife", "person"]:
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show processed video frame using cv2_imshow for Colab
    cv2_imshow(frame) # Use cv2_imshow instead of cv2.imshow

    # Exit when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
